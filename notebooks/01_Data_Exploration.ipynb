```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("All libraries imported!")
```

```python
# Load the Dataset
print("Loading dataset... wait a moment...")

df = pd.read_csv(NLP_PATH + 'consumer_complaints.csv', low_memory=False)

print("="*50)
print("DATASET LOADED!")
print("="*50)
print(f"Total rows:    {len(df):,}")
print(f"Total columns: {len(df.columns)}")
print(f"\nAll column names:")
for col in df.columns:
    print(f"  - {col}")
```

```python
print("FIRST 5 ROWS:")
df.head()
```

```python
print("DATA TYPES:")
print(df.dtypes)
```

```python
print("MISSING VALUES:")
missing = df.isnull().sum()
print(missing[missing > 0])
```

```python
# Explore Product Categories
print("="*50)
print("PRODUCT CATEGORIES")
print("="*50)
print(df['product'].value_counts())

plt.figure(figsize=(12, 6))
df['product'].value_counts().head(10).plot(
    kind='bar',
    color='steelblue',
    edgecolor='black'
)
plt.title('Top 10 Complaint Categories', fontsize=14, fontweight='bold')
plt.xlabel('Product Category')
plt.ylabel('Number of Complaints')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig(NLP_PATH + 'category_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved to Drive!")
```

```python
# Check exact column name first
print(df.columns.tolist())

# Find the date column — handles any naming format
date_col = [col for col in df.columns if 'date' in col.lower() or 'Date' in col]
print(f"\nDate columns found: {date_col}")
```

```python
print("="*50)
print("PRODUCT CATEGORIES")
print("="*50)
print(df['product'].value_counts())

plt.figure(figsize=(12, 6))
df['product'].value_counts().head(10).plot(
    kind='bar', color='steelblue', edgecolor='black'
)
plt.title('Top 10 Complaint Categories', fontsize=14, fontweight='bold')
plt.xlabel('Product Category')
plt.ylabel('Number of Complaints')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig(NLP_PATH + 'category_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved!")
```

```python
# Since no date column, analyze complaint text length instead
df['text_length'] = df['narrative'].astype(str).apply(len)

plt.figure(figsize=(12, 4))
df['text_length'].clip(upper=5000).hist(bins=50, color='#e74c3c', edgecolor='black')
plt.title('Distribution of Complaint Text Length', fontweight='bold', fontsize=14)
plt.xlabel('Number of Characters')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig(NLP_PATH + 'text_length_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved!")
```

```python
# No company column in this dataset — show category breakdown instead
df['product'].value_counts().head(10).plot(
    kind='barh', color='#2ecc71', edgecolor='black', figsize=(12, 5)
)
plt.title('Top Categories (Horizontal)', fontweight='bold', fontsize=14)
plt.xlabel('Count')
plt.tight_layout()
plt.savefig(NLP_PATH + 'top_categories_horizontal.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved!")
```
